### Implementation: Define a Performance Metric
It is difficult to measure the quality of a given model without quantifying its performance over training and testing. This is typically done using some type of performance metric, whether it is through calculating some type of error, the goodness of fit, or some other useful measurement. For this project, you will be calculating the [*coefficient of determination*](http://stattrek.com/statistics/dictionary.aspx?definition=coefficient_of_determination), R<sup>2</sup>, to quantify your model's performance. The coefficient of determination for a model is a useful statistic in regression analysis, as it often describes how "good" that model is at making predictions. 

The values for R<sup>2</sup> range from 0 to 1, which captures the percentage of squared correlation between the predicted and actual values of the **target variable**. A model with an R<sup>2</sup> of 0 is no better than a model that always predicts the *mean* of the target variable, whereas a model with an R<sup>2</sup> of 1 perfectly predicts the target variable. Any value between 0 and 1 indicates what percentage of the target variable, using this model, can be explained by the **features**. _A model can be given a negative R<sup>2</sup> as well, which indicates that the model is **arbitrarily worse** than one that always predicts the mean of the target variable._

For the `performance_metric` function in the code cell below, you will need to implement the following:
- Use `r2_score` from `sklearn.metrics` to perform a performance calculation between `y_true` and `y_predict`.
- Assign the performance score to the `score` variable.

```python
# TODO: Import 'r2_score'
from sklearn.metrics import r2_score

def performance_metric(y_true, y_predict):
    """ Calculates and returns the performance score between 
        true and predicted values based on the metric chosen. """
    
    # TODO: Calculate the performance score between 'y_true' and 'y_predict'  
    score = r2_score(y_true, y_predict)
    
    # Return the score
    return score
```

### Question 7 - Grid Search
*What is the grid search technique and how it can be applied to optimize a learning algorithm?*

**Answer: **
The machine learning algorithms we use have different hyperparameters which control different aspects over exactly how the work. A good example is the max_depth parameter for the Decision Tree algorithm that we used so far in this project which sets an upper bound on the number of levels of hierachy in the tree.

When we want to apply a learning algorithm to a problem we have to set these hyperparameters, but it wont usually be clear exactly what the optimal parameters should be. It can be complicated enough just trying to tune one parameter as we saw so far in this project, but in reality we might want to set 2,3 or even more different parameters. Changing one parameter will almost certainly change the optimal value for one of the other ones. As the number of hyperparameters increases, the number of options increases exponentially, so we need a hystematic way to keep track of and determine the best parameters to use. This is where the grid search technique comes in.

In the grid search technique we tell the algorithm which n parameters we want to tune and a list of potential values. It makes a n-dimensional grid of all possible combinations of all parameters and then runs our target machine learning algorithm with each of those combinations, keeping track of the scores/errors obtained by each. It then selects the parameters which created the highest score/lowest error and presents these as the optimal tuning values for the machine learning algorithm.

### Question 8 - Cross-Validation
*What is the k-fold cross-validation training technique? What benefit does this technique provide for grid search when optimizing a model?*  
**Hint:** Much like the reasoning behind having a testing set, what could go wrong with using grid search without a cross-validated set?

**Answer: **
As already discussed it is important for us to split the data into training and testing sets. But when we do grid search we need further subdivide the data into training, validation and testing sets. The model is first created with the training set, then hyperparameters are tuned with the validation set and finally the models score evaluated with the testing set. But we have a difficult trade-off to make in determining how much data to allocate to each sub-set. The more data we use for training in theory the better the model we can make. But the more data we use for validation the better we can choose the hyperparameters. And finally the more data used for testing, the better we can evaluate how well the model actually worked. Splitting up the data into too small groups can lead to bad performance and this is especially important if we have only have a small data set to start off with.

Using k-fold cross-validation training technique can help avoid us having to make this difficult decision. We still have to separately set aside a sub-portion of the data for testing, but then k-folds helps us maximise the use of the remaining data to do both training and validation. K-folds splits the remaining data into k equally-sized subsets, or 'folds'. It then trains the model using k-1 of the folds and uses the remaining fold for the validation. It then repeats this process k times, each time using a different fold for the validation and the remaining folds for training. The final reported performance of the model is the average score across all k runs.  This increases computational complexity, but means all data is used at some point in both training and validation so maximises both the potential of creating the best possible model and having best hyperparameters. It avoids wasting data which is especially important when the number of samples is small.

If we did grid-search without having a cross-validated set we could end up overfitting the hyperparameters to the testing data rather than correctly modelling the underlying trends.
